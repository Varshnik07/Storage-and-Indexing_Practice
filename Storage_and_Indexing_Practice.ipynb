Create Enbeddings with **sentence-transformers**

```python
!pip install sentence-transformers

from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

texts = [
    "The student performance system analyzes learning patterns.",
    "Bloom's taxonomy is used for generating questions.",
    "We extract metadata from PDFs.",
]

embeddings = model.encode(texts)
print(embeddings.shape)  # (3, 384)
```

Store in **Chroma** (lightweight local vectorDB)

```python
!pip install chromadb
import chromadb
from chromadb.config import Settings

client = chromadb.Client(Settings())
collection = client.create_collection(name="project_docs")

# Insert documents
collection.add(
    documents=texts,
    embeddings=embeddings.tolist(),
    ids=["1", "2", "3"]
)
```

Performing a Similarity Search

```python
query = "How do we create questions for students?"
query_embed = model.encode([query]).tolist()

results = collection.query(
    query_embeddings=query_embed,
    n_results=2
)

print("\nTop Matches:")
for doc in results["documents"][0]:
    print("-", doc)
```
